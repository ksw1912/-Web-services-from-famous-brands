```python
import pandas as pd
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.inception_resnet_v2 import preprocess_input
from keras.models import load_model

#미리 학습된 데이터 로드(서버에 이식할 때 경로 수정!)
model = load_model("/content/drive/MyDrive/Learning_Project/Final.h5")

#데이터셋 클래스
class_dictionary = {
    "Apple_A" : 0,
    "Facebook_A" : 1,
    "Instagram" : 2,
    "Kakaotalk_A" : 3,
    "McDonald's_A" : 4,
    "McDonald's_B" : 5,
    "Nike_A" : 6,
    "Nike_B" : 7,
    "Porsche_A" : 8,
    "Porsche_B" : 9,
    "Tiktok_A" : 10,
}
#이미지 사이즈
IMAGE_SIZE = (200, 200)

#테스트 이미지              #테스트 이미지 경로(수정 필요!)
test_image = image.load_img("/content/drive/MyDrive/Test_image/McDonald's_test_2.png", target_size =IMAGE_SIZE )
test_image = image.img_to_array(test_image)

#테스트 이미지 전처리
test_image = test_image.reshape((1, test_image.shape[0], test_image.shape[1], test_image.shape[2]))
test_image = preprocess_input(test_image)

#예측
prediction = model.predict(test_image)

#출력
df = pd.DataFrame({'pred':prediction[0]})
df = df.sort_values(by='pred', ascending=False, na_position='first')
print(f"예측률 : {(df.iloc[0]['pred'])* 100:.2f}%")

for x in class_dictionary:
  if class_dictionary[x] == (df[df == df.iloc[0]].index[0]):
    print(f"Class prediction = {x[:-2]}")
    break
```
